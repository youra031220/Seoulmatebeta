```python
import gensim
import numpy as np
import pprint as pp
```

```python
import gensim.downloader
list(gensim.downloader.info()['models'].keys())
```

```python
your_model_code = 'glove-twitter-25' # select among the model code aboves 숫자 더큰거 하면 모델 커짐
model = gensim.downloader.load(your_model_code) # download and load the model. It can take some time
```

```python
your_word = 'computer'
model[your_word]
```

```python
model.most_similar(your_word)
```

```python
import matplotlib.pyplot as plt

plt.plot(model['apple'])
plt.plot(model['samsung'])
plt.plot(model['orange'])
plt.plot(model['coding'])
```
